## Introduction

This is a Jupyter notebook that contains an analysis of BTC-in-USD closing prices in the year 2023. The analysis prescribes closing prices going to the third quarter of 2023 (Q3) along with statistical metrics and visualizations.

Included in the notebook:
- Data loading and cleaning
- Statistical analysis (mean, median, standard deviation, etc.)
- Visualization of the Q3 data
- Identification of outliers and analysis of daily price changes. 

### To do the above:
1. Load and clean data
2. Analyze Q3 data (day 182 through day 273)
3. Plot the closing prices for Q3
4. Analyze and visualize price changes and outliers.

This notebook will provide insights into the price behavior of Bitcoin over the specified period.

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = np.loadtxt('BTC_CLOSING_PRICES_2023.csv', delimiter=',')
```

### Step 1: Statistical Analysis for Quarter 3 of 2023 (Days 182-273)
We will compute the statistics as follows during the third quarter of 2023- from day 182 to day 273:
- Arithmetic Mean
- Minimum
- First Quartile
- Median
- Third Quartile
- Maximum
- Standard Deviation
- IQR

```python
# Slice the data for Q3 (Days 182 to 273)
q3_data = data[181:273]  # This assumes day 182 is index 181 in Python (0-indexed)

# Compute the required aggregates
mean = np.mean(q3_data)
minimum = np.min(q3_data)
Q1 = np.percentile(q3_data, 25)
median = np.median(q3_data)
Q3 = np.percentile(q3_data, 75)
maximum = np.max(q3_data)
std_dev = np.std(q3_data)
IQR = Q3 - Q1

# Print the results
print("Arithmetic Mean:", mean)
print("Minimum:", minimum)
print("Q1 (First Quartile):", Q1)
print("Median:", median)
print("Q3 (Third Quartile):", Q3)
print("Maximum:", maximum)
print("Standard Deviation:", std_dev)
print("Interquartile Range (IQR):", IQR)
```

### Step 2: Graphic Representation of Q3 Data (from days 182 to 273)
Now, to understand the trends associated with the third quarter (?3) or, even more, in 2023, one should plot the BTC closing prices during the specific period.

```python
# Plot the Q3 data
days = np.arange(182, 274)  # Day numbers from 182 to 273
plt.plot(days, q3_data, color='red', linestyle='-', linewidth=2)

# Add titles and labels
plt.title("BTC-to-USD Closing Prices (Q3 2023)")
plt.xlabel("Day")
plt.ylabel("Price (USD)")

# Display the plot
plt.show()
```

### Step 3: Identify the Price Lowest and Highest Days in Q3 2023
Now, we are interested in finding the **lowest** and **highest** prices that were recorded for the period of Q3 2023.

```python
# Find the lowest and highest observed prices and their corresponding days
lowest_price_day = np.argmin(q3_data) + 182  # +182 to get the actual day number
highest_price_day = np.argmax(q3_data) + 182  # +182 to get the actual day number

lowest_price = np.min(q3_data)
highest_price = np.max(q3_data)

# Print the results
print(f"Lowest price was on day {lowest_price_day} ({lowest_price})")
print(f"Highest price was on day {highest_price_day} ({highest_price})")
```

### Step 4: Daily Price Change Box Plot (For Postgraduate)
Next up, we'll be making a boxplot for the price changes (the difference between the closing prices of two consecutive days).

```python
# Calculate daily price changes
daily_changes = np.diff(q3_data)

# Plot the box plot
plt.boxplot(daily_changes, vert=False, patch_artist=True)

# Mark the arithmetic mean on the box plot
mean_change = np.mean(daily_changes)
plt.plot(mean_change, 1, 'gx')  # Green 'x' for the mean

# Add labels and title
plt.title("Box-and-Whisker Plot of BTC-to-USD Daily Price Changes (Q3 2023)")
plt.xlabel("Price Change (USD)")
plt.show()
```

### Step 5: Count Outliers in Box Plot
We are going to count the number of outliers in daily price changes using the Interquartile Range (IQR) method.

```python
# Calculate the outliers using the IQR method
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = daily_changes[(daily_changes < lower_bound) | (daily_changes > upper_bound)]

# Print the results
print("Number of outliers:", len(outliers))
print("Outliers:", outliers)
```
